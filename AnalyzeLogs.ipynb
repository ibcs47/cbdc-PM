```python
import pm4py
import json
import pandas as pd
```

```python
with open('eventLogsJSON.json','r') as f:
    d = json.load(f)

df = pd.json_normalize(d['values'])
df['address'].fillna(df['from'], inplace=True)
df.head(20)
```

```python
event_log = pm4py.format_dataframe(df, case_id='address',
                                       activity_key='name', timestamp_key='timestamp')
```

```python
process_tree = pm4py.discover_process_tree_inductive(event_log)
bpmn_model = pm4py.convert_to_bpmn(process_tree)
pm4py.view_bpmn(bpmn_model)
```

```python
net, im, fm = pm4py.discover_petri_net_heuristics(event_log)
pm4py.view_petri_net(net, im, fm)
```

```python
from pm4py.algo.discovery.alpha import algorithm as alpha_miner
from pm4py.algo.discovery.inductive import algorithm as inductive_miner
from pm4py.algo.discovery.heuristics import algorithm as heuristics_miner

from pm4py.visualization.petri_net import visualizer as pn_visualizer
```

```python
log = event_log
miner = "alpha"

if(miner=="alpha"):
    net, im, fm = alpha_miner.apply(log)
elif(miner=="inductive"):
    net, im, fm = inductive_miner.apply(log)
elif(miner=="heuristics"):
    net, im, fm = heuristics_miner.apply(log, parameters={heuristics_miner.Variants.CLASSIC.value.Parameters.DEPENDENCY_THRESH: 0.99})


gviz = pn_visualizer.apply(net, im, fm)
pn_visualizer.view(gviz)
```

```python
#Inductive
net, im, fm = inductive_miner.apply(log)
gviz = pn_visualizer.apply(net, im, fm)
pn_visualizer.view(gviz)
```

```python
#Heuristics
net, im, fm = heuristics_miner.apply(log, parameters={heuristics_miner.Variants.CLASSIC.value.Parameters.DEPENDENCY_THRESH: 0.99})

gviz = pn_visualizer.apply(net, im, fm)
pn_visualizer.view(gviz)
```

```python
df2 = df
df2["case_id"] = str(df2["address"])[0:14] + str(df2["timestamp"])[0:21]
df2.head(2)
```
